#Label Encoding

```python
from sklearn.preprocessing import LabelEncoder       #Importing LabelEncoder from scikit learn library.
le = LabelEncoder()                                  #Creating an object for LabelEncoder()
```

```python
'''Label Encoding the 'sex' column in df DataFrame and it contains string type value(categorical-Dichotomous),
#so it needs to be encoded.'''

df.sex = le.fit_transform(df[['sex']])
```

```python
df.head(7)    
```

```python
#'age' and 'bmi' columns contain numeric type of data,So it does not need to be encoded. After encoding the 'sex' column, 
#Now the 'smoker' column will be encoded because it contains yes/no type value(Categorical-Dichotomous variable)

df.smoker = le.fit_transform(df[['smoker']]) 
```

```python
df.head()
```

```python
df.region = le.fit_transform(df[['region']])    # The region column contains Categorical-Nominal variable so encoding required.
```

#One Hot Encoding

```python
dummy0 = pd.get_dummies(df1.sex, drop_first = True, prefix = 'sex')     

# Creating dummy variables from unique values of 'sex' column that is, 'female','male' - 2 separate columns will be created.
# After getting dummy variables and their respective column, drop the first column so,between 2 columns only one column remains.
# prefix = 'sex', so it will appear as, sex_male, sex_female.
```

```python
dummy0.head()       # Female column has been dropped.
```

```python
df1.drop('sex', axis = 1, inplace = True)      # Older 'sex' column is of no use, so it is dropped from df1 
                                            # inplace = True means, we are dropping the older'sex' column from df1 permanently.
```

```python
df1.head()
```

```python
# The new dummy variable column added with df1 from which older 'sex' column has been deleted.
# concat means adding.

df1 = pd.concat([df1, dummy0], axis = 1)   
```

```python
df1.head()
```

```python
# Getting dummy variables from 'smoker' column means, smoker_no, smoker_yes, Dropping the first column between two columns.

dummy1 = pd.get_dummies(df1.smoker,drop_first=True,prefix = 'smoker')  
```

```python
dummy1.head()
```

```python
df1.head()
```

```python
df1.drop('smoker', axis = 1, inplace = True)    # Deleting older'smoker' column permanently bcz of no use of it.
```

```python
df1.head()
```

```python
df1 = pd.concat([df1,dummy1], axis = 1)       # Adding the dummy variables column with df1 from which 'smoker' column deleted.
```

```python
df1.head()
```

```python
dummy2 = pd.get_dummies(df1.region, drop_first = True, prefix = 'region') #Getting dummies from 'region' column.
```

```python
order0 = df5.sex.unique()          #'order0' variable contains the unique values of 'sex' column and we can set the order.
order0                             #order0 = ['male','female']
```

```python
ordinal = OrdinalEncoder(categories=[order0])      # Creating an object for OrdinalEncoder.
```

```python
ordinal                           # By default sequence, 'female' is first, and 'male' second.
```

```python
encoded = ordinal.fit_transform(df5[['sex']])    #Ordinal encoding is done on 'sex' column and kept inside 'encoded' variable.
```
