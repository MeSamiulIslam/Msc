```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/MeSamiulIslam/Msc/main/Cost_of_Living_Index_2022.csv"
df = pd.read_csv(url)
```

```python
df.head()
```

```python
df.head(3)
```

```python
df.count()
```

```python
#Renaming Columns with inplace
df.rename(columns = {"Cost of Living Index": "Cost of Living"},inplace=True) 
```

```python
df.head(2)
```

```python
#Check Null values
df.isnull().values.any()
```

```python
#Count the number of Null values
df.isnull().values.sum()
```

```python
#Showing the null values in all attributes
null_columns = df.columns[df.isnull().any()]
df[null_columns].isnull().sum()
```

```python
# Showing where the  Null Values exist with row 
print(df[df.isnull().any(axis=1)][null_columns].head())
```

```python
df.head()
```

```python
#Dropping columns
df.drop(['Rank'], axis=1)
```

```python
#dropping rows
df.drop([8,18,57])
```

```python
# Filling null values with specific value
df['Cost of Living'].fillna(25.50,inplace=True)
df
```

```python
df.count()
```

```python
# Row wise data filling with loc function
df.loc[88,'Groceries Index'] = 55
df.head()
```

```python
print(df[df.isnull().any(axis=1)][null_columns].head())
```

```python
#filling using mode
df['Groceries Index'].fillna(df['Groceries Index'].mode()[0], inplace=True)
df
```

```python
print(df[df.isnull().any(axis=1)][null_columns].head())
```

```python
#Group By parameter check
#df.groupby('Rent Index')['Cost of Living'].median()
```

```python
#fill using "bfill" method
df['Rent Index']=df['Rent Index'].fillna(method='bfill')
df
```

```python
#Renaming Columns with inplace
df.rename(columns = {"Cost of Living Index": "Cost_of_Living"},inplace=True) 
df.rename(columns = {"Rent Index": "Rent_Index"},inplace=True) 
df.rename(columns = {"Cost of Living Plus Rent Index": "Cost_of_Living_Plus_Rent"},inplace=True) 
df.rename(columns = {"Groceries Index": "Groceries"},inplace=True) 
df.rename(columns = {"Restaurant Price Index": "Restaurant_Price"},inplace=True) 
df.rename(columns = {"Local Purchasing Power Index": "Local_Purchasing_Power"},inplace=True) 
```
